## Death count analysis in Germany from 2016 to 2022

Germany, officially the Federal Republic of Germany, is a country in Central Europe. It is the second most populous country in Europe after Russia, and the most populous member state of the European Union. Germany is situated between the Baltic and North seas to the north, and the Alps to the south; it covers an area of 357,022 square kilometres (137,847 sq mi), with a population of over 83 million within its 16 constituent states. Let's investigate and analyse the death count in Germany and it's states.


First, we analyse the Crude Death Count, Average Daily Death Count, and Total Death Count across different age groups and gender in Germany from 2016 to 2022.

```python
#importing modules and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the dataset to be used
crude_df = pd.read_excel('total daily deaths in Germany.xlsx')
crude_df.head()

```

```python
#renaming the necessary columns
crude_df = crude_df.rename(columns={crude_df.columns[0]: 'Year', crude_df.columns[-1]: 'Total'})
crude_df
```

```python
#the datset evidently contains null values hence we fill them with 0 values.
crude_df.fillna(value=0)
```

```python
#computing the sum of daily deaths inthe year 2022
a = crude_df.iloc[:1,1:]
a.sum(axis=1)
```

```python
#subsetting the total number of deaths
crude_dfn = crude_df[['Year', 'Total']]
crude_dfn
```

```python
#summation on the total death count
crude_dfn.iloc[:1,1:] = a.sum(axis=1)
```

```python
crude_dfn
```

```python
#plotting the total number of deaths over the 7-year period

Ydf =  crude_dfn.Year
Totall = crude_dfn.Total
plt.subplots(figsize=(10,8))
plt.grid(True, linestyle='-')
plt.plot(Ydf,Totall, marker='*', color='green', ms=10)
plt.bar(Ydf,Totall)
plt.xticks(rotation = 90)
plt.title("Crude Death Count in Germany from 2016 to 2022")
plt.xlabel('Year')
plt.ylabel("Number of deaths")
plt.style.use('tableau-colorblind10')
plt.show()
print("The drastic drop in 2022 is due to lack of data records between August and December 2022 at the time of this analysis.")
```

Evaluating the total death count over the 7-year period can be misleading as there's not enough data taken for the year 2022 hence the drastic drop we see. To remedy this and get results over continuous data for the 7-year period, we use the average daily death count.

```python
#importing the required dataset
df_for_av = pd.read_excel('total daily deaths in Germany.xlsx')
df_for_av
```

```python
#renaming the needed columns
df_for_av = df_for_av.rename(columns={df_for_av.columns[0]: 'Year'})
df_for_av
```

```python
#subsetting values needed to compute the average
df_a = df_for_av.iloc[:,1:-1]
df_a
```

```python
#evaluating the average death count in the year 2022
df_22 = df_for_av.iloc[:1,1:214].mean(axis=1)
df_22
```

```python
#evaluating the average death count over the 7-year period
df_a['average'] = df_a.mean(axis=1).dropna()
df_a
```

```python
#plotting the average daily death count in germany of the 7-year period
m =  df_for_av.Year
n = df_a.average
plt.subplots(figsize=(12,8))
plt.grid(True, linestyle='--')
plt.plot(m,n, marker='*', color='black', ms=10)
plt.bar(m,n, color='red')
plt.xticks(rotation = 90)
plt.title("Average Daily Death Count in Germany from 2016 to 2022")
plt.xlabel('Year')
plt.ylabel("Number of deaths")
plt.show()
```

We now see the trend in death count is much more uniform. We can also observe that the average daily death count has been on a steady increase from 2016 to 2022.

We see that the Average Monthly Death count between both genders has been on an overall increase over the last 5 years. Both male and female death counts had a uniform increase in 2016, 2017, and 2018 with female count being higher. Both death counts significantly decrease in 2019 and then experience a great increase in 2020 where both counts are equal and 2021 where the male death count became higher than female death count. There is a decrease in male death count and a slight increase in female death count so far in 2022.
